```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("Canada_neighborhoods.csv", sep=";")
```

```python
df.head()
```

```python
df1 = df[df.Borough != "Not assigned"]
df1.head(10)
```

```python
df2 = df1.copy()
df2 = df2.groupby(['Postcode', 'Borough']).Neighbourhood.agg([('Neighbourhood', ', '.join)]).reset_index()
df2.head()
```

```python
df3 = df2.copy()
df3.loc[df3['Neighbourhood'] == "Not assigned", "Neighbourhood"] = df3.loc[df3['Neighbourhood'] == "Not assigned", "Borough"]
df3.head(10)
```

```python
df3.to_csv("Canada_neighborhoods_clean.csv", sep=";")
```
